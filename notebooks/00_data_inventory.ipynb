```python
import pandas as pd
from pathlib import Path

data_dir = Path("../data/raw")

print("Files found in data/raw:")
for f in data_dir.iterdir():
    print("-", f.name)

# Load data_core.csv
core_path = data_dir / "data_core.csv"
core_df = pd.read_csv(core_path)
print("\n=== data_core.csv ===")
print("Shape:", core_df.shape)
display(core_df.head())

# Load India Agriculture Crop Production.csv
prod_path = data_dir / "India Agriculture Crop Production.csv"
prod_df = pd.read_csv(prod_path)
print("\n=== India Agriculture Crop Production.csv ===")
print("Shape:", prod_df.shape)
display(prod_df.head())
```
